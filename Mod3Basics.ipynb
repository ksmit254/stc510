```python
import pandas as pd
import csv
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('crimestat.csv', header = None, delimiter=',', quoting=csv.QUOTE_NONE, encoding='utf-8')
```

```python
df.rename(columns={0: 'inc_num', 1: 'occured_on', 2: 'occured_to', 3: 'crime_cat', 4: 'address', 5: 'zip', 6: 'housing_type'}, inplace=True)
```

```python
df_reset=df.set_index('inc_num',inplace=True)
```

```python
df.drop(labels='"INC NUMBER"', axis=0,inplace=True)
```

```python
df.head()
```

```python
df.sort_index(inplace=True)
```

```python
df.sort_values(by='housing_type')
```
